#### 9. Now, replace the missing values in DataFrame

```python
for colname in temp_aus_df.columns:
    if temp_aus_df[colname].dtype == 'object':
        temp_aus_df[colname] = temp_aus_df[colname].fillna('others')
    else:
        temp_aus_df[colname] = temp_aus_df[colname].fillna(0)
```
